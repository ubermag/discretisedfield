## Initiate `VertexField`

```python
import discretisedfield as df
```

```python
p1 = (-50, -50, -50)
```

```python
p2 = (50, 50, 50)
```

```python
mesh = df.Mesh(p1=p1, p2=p2, n=(100, 100, 100))
```

```python
def Ms(point):
    x, y, z = point
    if x**2 + y**2 <= (20) ** 2:
        return 1
    else:
        return 0
```

```python
vertex_field = df.VertexField(mesh, nvdim=3, value=[0, 0, 1], norm=Ms)
```

```python
vertex_field.hv(kdims=["x", "y"])
```

```python
vertex_field_xrr = vertex_field.to_xarray()
```
