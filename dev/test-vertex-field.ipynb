## Initiate `VertexField`

```python
import discretisedfield as df
```

```python
p1 = (0, 0, 0)
```

```python
p2 = (100, 100, 100)
```

```python
mesh = df.Mesh(p1=p1, p2=p2, n=(100, 100, 100))
```

```python
vertex_field = df.VertexField(mesh, nvdim=3, value=[0, 0, 1])
```

```python
vertex_field.to_xarray()
```

```python
vertex_field.hv(kdims=["z", "y"])
```

```python
vertex_field_xrr = vertex_field.to_xarray()
```
