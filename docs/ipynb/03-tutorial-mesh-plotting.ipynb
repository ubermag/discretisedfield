# Tutorial 3: Mesh plotting

There are two ways how a mesh can be plotted. Using

- `matplotlib` (`discretisedfield.Mesh.mpl`) or
- `k3d` (`discretisedfield.Mesh.k3d`)

For plotting, the inline plotting inside Jupyter notebook should be enabled.

```python
%matplotlib inline
```

A `matplotlib` representation of the mesh is

```python
import discretisedfield as df

p1 = (-3, -3, -3)
p2 = (3, 3, 3)
n = (6, 6, 6)
mesh = df.Mesh(p1=p1, p2=p2, n=n)

mesh.mpl()
```

Blue solid lines are the edges of the mesh domain, whereas the dashed red lines show the relative size of a single discretisation cell. To change the size of the plot, `figsize` parameter can be passed. It is a tuple of size 2.

```python
mesh.mpl(figsize=(6, 6))
```

In order to get an interactive 3D plot inside Jupyter notebook, `discretisedfield.Mesh.k3d` method can be used.

```python
mesh.k3d()
```
